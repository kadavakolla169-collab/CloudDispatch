```python
import matplotlib.pyplot as plt
import numpy as np

workers = [1, 2, 4]
throughput = [724, 3778, 5400]
latency = [0.299, 0.300, 0.300]

fig, ax1 = plt.subplots(figsize=(8, 5))

bars = ax1.bar(workers, throughput, color=['#B5D4F4','#378ADD','#185FA5'], width=0.6)
ax1.set_xlabel('Number of Worker VMs')
ax1.set_ylabel('Throughput (req/sec)', color='#185FA5')
ax1.set_xticks(workers)
ax1.set_xticklabels(['1 Worker', '2 Workers', '4 Workers'])
ax1.tick_params(axis='y', labelcolor='#185FA5')

ax2 = ax1.twinx()
ax2.plot(workers, latency, color='#1D9E75', marker='o', linewidth=2, markersize=8)
ax2.set_ylabel('Avg Latency (s)', color='#1D9E75')
ax2.set_ylim(0.295, 0.305)
ax2.tick_params(axis='y', labelcolor='#1D9E75')

plt.title('CloudDispatch Scaling Benchmark — Throughput vs Worker VMs')
plt.tight_layout()
plt.savefig('scaling_benchmark.png', dpi=150)
plt.show()
print("Chart saved as scaling_benchmark.png")
```

```python
import matplotlib.pyplot as plt

timeline = ['All 4 workers\nrunning', 'Worker-North\nkilled', '3 workers\ncontinue']
throughput = [5400, 4863, 4863]

plt.figure(figsize=(8, 5))
plt.plot(timeline, throughput, marker='o', linewidth=2.5,
         markersize=10, color='#378ADD')
plt.fill_between(range(3), throughput, alpha=0.1, color='#378ADD')

plt.axvline(x=1, color='red', linestyle='--', linewidth=1.5, label='Node failure')
plt.annotate('Worker-North\nfailed here', xy=(1, 5100),
             fontsize=10, color='red', ha='center')

plt.ylabel('Throughput (req/sec)')
plt.title('CloudDispatch Fault Tolerance — System Behavior on Node Failure')
plt.ylim(4000, 6000)
plt.legend()
plt.tight_layout()
plt.savefig('fault_tolerance.png', dpi=150)
plt.show()
print("Chart saved as fault_tolerance.png")
```
